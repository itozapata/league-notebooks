```python
import pandas as pd
import matplotlib.pyplot as plt

import plotly.offline as offline
import plotly.plotly as py
import plotly.graph_objs as go
import cufflinks as cf
from plotly.offline import iplot
print(pd.__version__)

offline.init_notebook_mode()
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('./datasets/league_games_by_lanes.csv')
df.head(5)
```

```python
pmint_summoners = [
    'Baaby Wipes ', 
    'Skeevo', 
    'dstrawyer28', 
    'Sheldor', 
    'Archadium', 
    'BokuWaHenry', 
    'AgileOverlord', 
    'jokokoloko', 
    'ChapJae', 
    'Channsanity', 
    'Rudo',
    'cyrst0lpeTe',
    'Onishfu'
]
game_limit = 5
```

```python
def filterBySummoners(val, summoners):
    if val in summoners:
        return val
    return 'Unknown'

df['pmint_top_summonerName'] = df['pmint_top_summonerName'].apply(lambda x: filterBySummoners(x, pmint_summoners))
df['pmint_mid_summonerName'] = df['pmint_mid_summonerName'].apply(lambda x: filterBySummoners(x, pmint_summoners))
df['pmint_jungle_summonerName'] = df['pmint_jungle_summonerName'].apply(lambda x: filterBySummoners(x, pmint_summoners))
df['pmint_adc_summonerName'] = df['pmint_adc_summonerName'].apply(lambda x: filterBySummoners(x, pmint_summoners))
df['pmint_support_summonerName'] = df['pmint_support_summonerName'].apply(lambda x: filterBySummoners(x, pmint_summoners))
df.head()
```

```python
df2 = df[['pmint_top_summonerName','pmint_mid_summonerName','pmint_jungle_summonerName','pmint_adc_summonerName','pmint_support_summonerName','pmint_win']].groupby(['pmint_top_summonerName','pmint_mid_summonerName','pmint_jungle_summonerName','pmint_adc_summonerName','pmint_support_summonerName'])
df2 = df2.filter(lambda x: len(x) >= game_limit)
df2 = df2.groupby(['pmint_top_summonerName','pmint_mid_summonerName','pmint_jungle_summonerName','pmint_adc_summonerName','pmint_support_summonerName'])
```

```python
df2.size().unstack().plot.barh(figsize=(10,10))
```

```python
df2.mean().unstack().plot.barh(figsize=(10,10))
```

```python
df2 = df[['pmint_summoners', 'pmint_win']].groupby(['pmint_summoners'])
df2 = df2.filter(lambda x: len(x) >= game_limit)
df2 = df2.groupby(['pmint_summoners'])
df2.head()
```

```python
df2.size()
```

# What is the collective experience of the Top Champion Pool

```python
games_with_pmint_top = df[df['pmint_top_summonerName'].isin(pmint_summoners)]
games_top = games_with_pmint_top.groupby(['pmint_top_summonerName', 'pmint_top']).filter(lambda x: len(x) >= game_limit)
```

#### How many games do we have on each champ and what is the distribution of who plays them?

```python
games_top.groupby(['pmint_top', 'pmint_top_summonerName']).size().unstack().plot.barh(figsize=(10,5), stacked=True)
```

#### Whats the win rate of these champions?

```python
games_top[['pmint_top','pmint_win']].groupby(['pmint_top']).mean().unstack().plot.barh(figsize=(10,5), stacked=True)
```

#### How many games do the players in this lane have and what is the distribution of their champ pool?

```python
games_top.groupby(['pmint_top_summonerName', 'pmint_top']).size().unstack().plot.barh(figsize=(10,5), stacked=True)
```
