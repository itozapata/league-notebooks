```python
def get_match_probability_json(match_id):
    conn = http.client.HTTPSConnection("api.sportradar.us")

    conn.request("GET", "/lol-t1/us/matches/{match_id}/probabilities.json?api_key={your_api_key}".format(your_api_key=api_key, match_id=match_id))

    res = conn.getresponse()
    data = res.read()    
    json_data = json.loads(data.decode('utf-8'))
    return json_data
```

```python
def get_match_from_file(match_id):
    path = "./prob/{id}.json".format(id=match_id.replace(":", "-"))
    json_data = json.load(open(path))
    return json_data
```

```python
def match_probability(match_id):
    json_data = get_match_from_file(match_id)
    
    df_prob = json_normalize(pd.DataFrame(json_data['probabilities'])['markets'])
    if df_prob.shape[0] == 0: 
        # one game didnt have a prediction so this takes care of that.
        return pd.DataFrame()
    
    df_prob = pd.DataFrame(df_prob['outcomes'][0])
    df_prob['qualifier'] = df_prob.name.str.split('_', expand=True)[[0]]
    df_competitors = pd.DataFrame.from_dict(json_data['sport_event']['competitors'])
    df_competitors['match_id'] = json_data['sport_event']['id']
    df_result = pd.merge(df_competitors, df_prob[['probability', 'qualifier']], on=['qualifier'])
    return df_result
```

# need to add in the actual results of the game next

```python
conn = http.client.HTTPSConnection("api.sportradar.us")

conn.request("GET", "/lol-t1/us/tournaments/sr:season:53672/results.json?api_key={your_api_key}".format(your_api_key=api_key))

res = conn.getresponse()
data = res.read()
```

```python
json_data = json.loads(data.decode('utf-8'))
json_data
df_match_results = json_normalize(data=json_data, record_path='results', meta=['tournament', 'generated_at', 'schema'])
df_match_results
df_match_results = pd.concat([df_match_results, df_match_results['sport_event_status'].apply(pd.Series)], axis = 1).drop('sport_event_status', axis = 1)
df_match_results = pd.concat([df_match_results, df_match_results['sport_event'].apply(pd.Series)], axis = 1).drop('sport_event', axis = 1)
df_match_results
```

```python
match_winners = df_match_results[['id', 'winner_id']]
match_winners.head()
```

```python
final_results = pd.merge(all_match_probs, match_winners, left_on=['match_id'], right_on='id')
final_results['expected_result'] = [1 if x.probability > 50 else 0 for i, x in final_results.iterrows()]
final_results['actual_result'] = [1 if x.id_x == x.winner_id else 0 for i, x in final_results.iterrows()]
final_results.to_parquet('./all_matches_with_predictions.snappy.parquet', compression='snappy')
final_results
```
