```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_parquet('./all_matches_with_predictions.snappy.parquet')
```

```python
df.head()
```

```python
df['diff'] = df.actual_result - df.expected_result
df['diffPct'] = (100*df.actual_result) - df.probability
df['diffPctOffEven'] = 50 - (100 - df.probability)

df.head()
```

```python
def funca(row):
    if row.probability >= 50 and row.actual_result == 1:
        return 1
    elif row.probability < 50 and row.actual_result == 0:
        return 1
    return 0
    
df['intersectPredictability'] = [funca(row) for i, row in df.iterrows()]
```

```python
a = df[['abbreviation', 'intersectPredictability']].groupby(['abbreviation'], as_index=False).mean()
b = pd.merge(left=df, right=a, on='abbreviation', suffixes=['_x',''])
b['conditionalPrediction'] = b.intersectPredictability * b.probability
df = b
```

```python
df.head()
```

```python
stats = df[['abbreviation', 'probability', 'conditionalPrediction', 'intersectPredictability']].groupby(['abbreviation'], as_index=False).agg({'probability': ['mean', 'std'], 'conditionalPrediction': ['mean', 'std'], 'intersectPredictability': ['max']})
stats
stats_reset = stats.reset_index()

new_df = pd.DataFrame()
new_df['team'] = stats_reset['abbreviation']

new_df['prob.mean'] = stats_reset['probability']['mean']
new_df['prob.std'] = stats_reset['probability']['std']
new_df['conditionalPrediction.mean'] = stats_reset['conditionalPrediction']['mean']
new_df['conditionalPrediction.std'] = stats_reset['conditionalPrediction']['std']
new_df['intersectPredictability'] = stats_reset['intersectPredictability']['max']
new_df['stdDiff'] = new_df['prob.std'] - new_df['conditionalPrediction.std']
new_df['meanDiff'] = new_df['prob.mean'] - new_df['conditionalPrediction.mean']
new_df
```

```python
# sns.set_context('talk')

p1 = sns.lmplot(x='intersectPredictability', y='conditionalPrediction.std', hue='team', fit_reg=False, data=new_df, scatter_kws={"marker": "D", "s": 120})
plt.title("2018 NA LCS Summer Split")
plt.xlabel("<- underperforming (mean prediction diff) overperforming ->")
plt.ylabel("<- predictable (std) unpredictable ->")


for index, row in new_df.iterrows():
    p1.ax.annotate(row['team'], (row['intersectPredictability'] + 0.005, row['conditionalPrediction.std']))

p1
```
